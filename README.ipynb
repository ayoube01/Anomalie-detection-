## About

Those notebooks are mainly about performing statistical anomaly detection against software performance indicators.

Several complimentary strategies exist.

https://blog.statsbot.co/time-series-anomaly-detection-algorithms-1cef5519aef2

### Outlier detection

Outlier detection can be used to flag strong spikes and dips in a time series as anomalies.

We make the following assumptions:
- our time series have a **normal distribution**
- **every outlier is a genuine anomaly** and not an error, given that our indicators come from software and not a human sensus

Still, we consider short-lived outliers (small spikes) as a normal behaviour that can have several low-level origins:
- garbage collection
- small contentions (network, disk ...)

We don't want to consider those as genuine incidents (at least if they're isolated).
Thus, we want to consider as **incidents** anomalies that occur for a **relatively extended** period of time (say more than 10s).

Moreover, most outlier detection techniques assume that the time series are relatively flat.
In case of seasonal data with strong variance across a period, we have to **wisely choose a window** that would reduce this variance.

### STL (Seasonal-Trend Decomposition)

Most of our time series are seasonal.

As such, to increase accuracy we can extract the seasonal component of the time series, giving us a flat time series on which we can do classical outlier detection.

Please see [Atlas.ipynb](./Atlas.ipynb).
